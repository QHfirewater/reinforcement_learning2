## 激活函数

### Sigmoid函数

```python
import torch
from matplotlib import pyplot as plt
```

```python
a = torch.linspace(-10,10,100)
a
```

```python
b = torch.sigmoid(a)
```

```python
plt.plot(a,b)
```

```python
from torch.nn import functional as f
```

```python
f.sigmoid(a)
```

### tanh

```python
a = torch.linspace(-5,5,1000)
a
```

```python
b = torch.tanh(a)
b
```

```python
plt.plot(a,b)
```

### ReLU函数

```python
a = torch.linspace(-5,5,100)
a
```

```python
b = torch.relu(a)
b
```

```python
plt.plot(a,b)
```

```python
c = f.relu(a)
```

```python
plt.plot(a,c)
```
