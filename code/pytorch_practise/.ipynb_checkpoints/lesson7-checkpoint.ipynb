## 感知机

### 单层感知机

```python
import torch 
from torch.nn import functional as F
```

```python
x = torch.randn(1,10)
w = torch.randn(1,10,requires_grad = True)
```

```python
o = torch.sigmoid(x@w.t())
o.shape
```

```python
loss = F.mse_loss(torch.ones(1,1),o)
loss.shape
```

```python
loss
```

```python
loss.backward()
```

```python
w.grad
```

### 多层的感知机

```python
x = torch.randn(1,10)
w = torch.randn(2,10,requires_grad = True)
```

```python
o = torch.sigmoid(x@w.t())
o.shape
```

```python
o
```

```python
loss = F.mse_loss(torch.ones(1,2),o)
loss
```

```python
loss.backward()
```

```python
w.grad
```

```python
w.shape
```
