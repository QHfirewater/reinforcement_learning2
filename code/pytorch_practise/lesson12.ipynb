## 全连接层

```python
import torch
from torch.nn import functional as F
from torch import nn
```

```python
x = torch.randn(1,784)
x.shape
```

```python
layer1 = nn.Linear(784,200)
layer2 = nn.Linear(200,200)
layer3 = nn.Linear(200,10)
```

```python
x = layer1(x)
x = F.relu(x)
x.shape
```

```python
x = layer2(x)
x = F.relu(x)
x.shape
```

```python
x = layer3(x)
x = F.relu(x)
x.shape
```

```python
x
```

```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
        nn.Linear(784,200),
        nn.ReLU(inplace =True),
        nn.Linear(200,200),
        nn.ReLU(inplace =True),
        nn.Linear(200,10),
        nn.ReLU(inplace =True))
        
    def forward(self,x):
        x = self.model(x)
        return x
```

```python
x = torch.randn(1,10)
```

```python
x = F.relu(x,inplace = True)
x
```

```python
x = torch.randn(1,10)
```

```python
layer = nn.ReLU()
x = layer(x)
x
```

```python
model = MLP()
```
