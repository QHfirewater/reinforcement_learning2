```python
import torch
from torch.nn import functional as F
import numpy as np
from matplotlib import pyplot as plt
```

```python
def himfun(x):
    y = (x[0]**2 +x[1]-11)**2 + (x[0]+x[1]**2-7)**2
    return y
```

```python
x = np.arange(-6,6,0.1)
y = np.arange(-6,6,0.1)
```

```python
x_,y_ = np.meshgrid(x,y)
z = himfun([x_,y_])
```

```python
x_.shape
```

```python
fig = plt.figure('himfun')
ax = fig.add_subplot(projection ='3d')
ax.plot_surface(x_,y_,z)
# ax.view_init(60,-20)
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
x = torch.tensor([4.,0.],requires_grad = True)
optimizer = torch.optim.Adam([x],lr = 1e-3)
```

```python
for step in range(20000):
    pred = himfun(x)
    optimizer.zero_grad()
    pred.backward()
    optimizer.step()
    if step %1000 ==0:
        print('step {}: x ={},f(x) = {}'.format(step,x.tolist(),pred.item()))
```
